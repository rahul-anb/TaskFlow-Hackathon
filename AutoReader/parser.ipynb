```python
setNodes_split = setNodes[setNodes.find("[") + 1: setNodes.rfind("]")]
print(setNodes_split)

template1 = {
                "id": "", 
                "type": "", 
                "data": { "label": "" }, 
                "position": {"x": 0, "y": 0}, 
                "sourcePosition":""
            }
template2 = {
                "id": "", 
                "type": "", 
                "data": {"onChange": "", "title":"", "deadline": "", "personAssigned": ""}, 
                "style": "", 
                "position": {"x": 0, "y": 0}
            }

dic = {}

for i in setNodes_split:
    # check for '\n'
    if i == '\n':
        print("new line")
    print(i,end="")
```

```python
setNodes_split1 = setNodes_split[setNodes_split.find("{"):]
# for i in setNodes_split1:
#     # check for '\n'
#     if i == '\n':
#         print("new line")
#     print(i,end="")

list_set_nodes=[]
# split after every 7th new line character
count = 0
start = 0
for i in range(len(setNodes_split1)):
    if setNodes_split1[i] == '\n':
        count += 1
        if count == 7:
            # print(i)
            list_set_nodes.append(setNodes_split1[start:i].strip())
            start = i+1
            count = 0

print(list_set_nodes)
```

```python
for i in list_set_nodes[0].split("\n"):
    val = i.strip()
    print(val)
    if val[:4]=='"id"':
        # print("found")
        template1["id"] = val[7:val.find('"',8)].strip()
    
    elif val[:6]=='"type"':
        template1["type"] = val[9:val.find('"',10)].strip()
        
    elif val[:6]=='"data"':
        data_inside = val[val.find("{")+1:val.rfind("}")].strip()
        data_inside = data_inside.split(",")
        
        # strip white spaces for all elements in list
        data_inside = [i.strip() for i in data_inside]    
        
        # print("sdfasd ",data_inside)
        for i in data_inside:
            # print(i[:7])
            if i[:7]=='"label"':
                # print("found")
                # print(i[9:i.find('"',10)].strip())
                template1["data"]["label"] = i[10:i.find('"',10)].strip()
    
    elif val[:10]=='"position"':
        data_inside = val[val.find("{")+1:val.rfind("}")].strip()
        data_inside = data_inside.split(",")
        data_inside = [i.strip() for i in data_inside] 
        # print("sdfasd ",data_inside)
        for i in data_inside:
            # print(i)
            if i[:3]=='"x"':
                # print(i[:3])
                template1["position"]["x"] = int(i[4:].strip())
                
            elif i[:3]=='"y"':
                # print(i[:3])
                template1["position"]["y"] = int(i[4:].strip())
    
    elif val[:16]=='"sourcePosition"':
        template1["sourcePosition"] = val[19:val.find('"',20)].strip()
        
        

print(template1)
```
